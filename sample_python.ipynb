21.	 Python - Write a python program to print the number which is missing in a list ranging numbers 1 to 100.


```python
def missing_list(lst):
    new_list = []
    for i in range(lst[0], lst[-1] + 1):
        if i not in lst:
            new_list.append(i)
    return new_list


def missing_list2(lst):
    list1 = [x for x in range(lst[0], lst[-1] + 1) if x not in lst]
    return list1


if __name__ == '__main__':
    lst = [1, 4, 5, 6, 7, 9]
    print(missing_list(lst))
    print(missing_list2(lst))
```

23.	Python - Given variable x='aaabbadcaa'
Write code to get the result like â€“ a3b2a1d1c1a2


```python
from collections import OrderedDict


def count_of_char(input_str):
    dict = OrderedDict.fromkeys(input_str, 0)
    for ch in input_str:
        dict[ch] = dict[ch] + 1

    output = ''
    for key, value in dict.items():
        output = output + key + str(value)
    return output


if __name__ == '__main__':
    input_str = 'aaabbadcaa'

    print(count_of_char(input_str))
```

```python
# single inheritance
# https://www.youtube.com/watch?v=u9x475OGj_U
class Parent:
    # create a class
    def first(self):
        print('this is parent class')


#   SINGLE INHERITANCE

# create second class and mention parent class in that
class Child(Parent):

    def second(self):
        print("this is child class")


#         create the object of child
obj1 = Child()
obj1.first()
obj1.second()
```

```python
# multiple
# https://www.youtube.com/watch?v=u9x475OGj_U

class Parent1:
    # create a class
    def first(self):
        print('this is parent1 class')


class Parent2:
    def first(self):
        print('this is parent2 class')

# if we have method with same name in both the class
# then the method from the parent will be called which is mention first in child class
# if we mention child class as -  class Child(Parent2, Parent1) then method from parent2 will be called
# if we mention class Child( Parent1, Parent2) then the method from parent 1 will be called
class Child(Parent2, Parent1):
    def second(self):
        print("this is child class")


obj1 = Child()
obj1.first()
obj1.second()
# obj1.third()
```

```python
# multi level
# https://www.youtube.com/watch?v=u9x475OGj_U
class Parent1:
    # create a class
    def first(self):
        print('this is parent1 class')


class Parent2(Parent1):
    def first(self):
        print('this is parent2 class')


class Child(Parent2):
    def second(self):
        print("this is child class")


obj1 = Child()
obj1.first()
obj1.second()
```

```python
# hybrid
# https://www.youtube.com/watch?v=u9x475OGj_U
# hybrid inheritance is the combination of two inheritance

class Parent1:
    # create a class
    def first(self):
        print('this is parent1 class')

# here we are using single inheritance
class Parent2(Parent1):
    def third(self):
        print('this is parent2 class')

# here we are using multiple inheritance
class Child(Parent2, Parent1):
    def second(self):
        print("this is child class")


obj1 = Child()
obj1.first()
obj1.second()
obj1.third()
```

```python
# poly
class Employee:
    # when same function behave diff in diff condition depend on aurgement pass
    # this can be achive using method overloading and method overriding

    def emp_name(self, name=''):
        return 'emp class name is :' + name


class Department(Employee):
    # because we have same method name with same argument in the  both classes
    # the object of Department is created so that method from Department will be called
    # method overriding
    def emp_name(self, name=''):
        return "dep. class" + name


obj = Department()
print(obj.emp_name())
print(obj.emp_name('abc'))
```

```python
# super function
# https://www.youtube.com/watch?v=u9x475OGj_U
# super function is use to call the parent method directly into chile

class Parent:
    # create a class
    def first(self):
        print('this is parent class')


#   SINGLE INHERITANCE

# create second class and mention parent class in that
class Child(Parent):

    def second(self):
        super().first()
        print("this is child class")


#         create the object of child
obj1 = Child()
# obj1.first()
obj1.second()
```
